```python
product_id=[1,2,3,4,5,6,7,8]
online_price=[233,456,770,120,222,444,200,300]
offline_price=[400,300,100,333,444,555,565,899]
discount_price=[100,200,300,400,150,250,450,500]
```

```python
plt.plot(product_id, online_price)
```

```python
plt.plot(product_id, online_price)
plt.xlabel('Product')
plt.ylabel('Prices')
plt.title('Product Price Comparision')
```

```python
plt.plot(product_id, online_price, color='red')
plt.xlabel('Product')
plt.ylabel('Prices')
plt.title('Product Price Comparision')
```

```python
plt.plot(product_id, online_price, color='red')
plt.xlabel('Product')
plt.ylabel('Prices')
plt.title('Product Price Comparision')
plt.show()
```

```python
plt.plot(product_id, online_price, color='red', marker='D', markersize=10)
plt.xlabel('Product')
plt.ylabel('Prices')
plt.title('Product Price Comparision')
plt.show()
```

```python
plt.plot(product_id, online_price, color='red', marker='D', markersize=10, label ='online_price')
plt.plot(product_id, offline_price, color='yellow', markersize=10, label ='offline_price')
plt.plot(product_id, discount_price, color='green', marker='+', markersize=10, label ='discount_price')
plt.xlabel('Product')
plt.ylabel('Prices')
plt.title('Product Price Comparision')
plt.legend()
plt.show()
```

```python
year=[2012,2013,2014,2015,2016,2017,2018]
price=[65000,69000,75000,55000,52000,64000,81000]
plt.bar(year, price, color='red')
plt.xlabel('Year')
plt.ylabel('Price')
plt.title('Year by Year Price comparision')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.style.use('fivethirtyeight')
```

```python
data=sns.load_dataset('tips')
```

```python
data.head()
```

Plot a Scatter Plot between total_bill and tip

```python
sns.relplot(x='total_bill', y='tip', data=data)
```

```python
sns.relplot(x='total_bill', y='tip', kind='scatter', data=data)
```

Hue Parameter

```python
sns.relplot(x='total_bill', y='tip', hue='smoker', data=data)
```

```python
sns.relplot(x='total_bill', y='tip', hue='smoker', style='sex',data=data)
```

Size parameter

```python
sns.relplot(x='total_bill', y='tip', hue='smoker', style='sex', size='size', sizes=(15,200), data=data)
```

```python
sns.scatterplot(x='total_bill', y='tip', hue='smoker', style='sex',data=data)
```

Categorical Scatterplots = Stripplots, Swarmplots

Heatmaps

```python
flights=sns.load_dataset('flights')
```

```python
flights.head()
```

```python
x=flights.pivot_table(index='year', columns='month', values='passengers', aggfunc='sum')
```

```python
sns.heatmap(x)
```

```python
sns.heatmap(x, cbar=False)
```

```python
sns.heatmap(x, cbar=False, linewidths=0.5)
```

```python
sns.heatmap(x, cbar=False, linewidths=0.5, annot=True, fmt='d')
```

```python
sns.heatmap(x, cbar=False, linewidths=0.5, annot=True, fmt='d', cmap='viridis')
```

```python
sns.heatmap(x, cbar=False, linewidths=0.5, annot=True, fmt='d', cmap='winter')
```

Clustermap

```python
sns.clustermap(x)
```

```python
sns.clustermap(x, metric='correlation')
```

```python
sns.clustermap(x, z_score=0, metric='correlation')
```

```python
sns.clustermap(x, z_score=0, annot=True, metric='correlation')
```

```python
sns.clustermap(x, z_score=0, annot=True, metric='correlation', row_cluster=False)
```

```python
sns.clustermap(x, z_score=0, annot=True, metric='correlation', col_cluster=False)
```

Distplot

```python
titanic=sns.load_dataset('titanic')
titanic.head()
```

```python
titanic['age'].fillna(titanic['age'].mean(), inplace=True)
```

```python
#DISTPLOT

sns.distplot(titanic['age'])
```

```python
#HISTOGRAM

sns.distplot(titanic['age'], kde=False)
```

```python
#HISTOGRAM

sns.distplot(titanic['age'], kde=False, bins=10)
```

```python
#KDE

sns.distplot(titanic['age'], hist=False)
```

```python
#RUGPLOT

sns.distplot(titanic['age'], hist=False, kde=False, rug=True)
```

```python
sns.distplot(titanic[titanic['survived']==1]['age'])
sns.distplot(titanic[titanic['survived']==0]['age'])
```
